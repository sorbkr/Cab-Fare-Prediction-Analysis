```python
#Concating two data sets.
a=pd.concat([cab_data,weather_data])
a.tail()
```

```python
#Taking the factor which affect the price to design our model.
a.columns
```

```python
x1=a[['distance','surge_multiplier','temp','clouds', 'pressure', 'rain','wind','day', 'hour']]
y1=a['price']
```

```python
#Using skicit learn to split data into tranning and testing sets 
#split the data into tranning and testing sets
x_train, y_train ,x_test, y_test = train_test_split(x1,y1,test_size=0.25,random_state=42)
linear=LinearRegression()
linear.fit(x_train,x_test)
```

```python
predictions=linear.predict(y_train)
predictions
```

```python
df =pd.DataFrame({'Actual': y_test,'Predicted': predictions})

#Plotting the graphs Actual vs. Predicted for better understanding of our model.
df1=df.head(25)
df1.plot(kind='bar',figsize=(26,10))
plt.grid(which='major',linestyle='-',linewidth='0.5',color='green')
plt.grid(which='minor',linestyle=':',linewidth='0.5',color='green')
plt.show()
```
